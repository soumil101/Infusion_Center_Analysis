```python
import pandas as pd
from datetime import datetime, time
```

```python
df = pd.read_csv('finalData.csv')
```

```python
def getTimeOfDay(dt):
    h = int(dt.hour)
    if h < 10:
        return 'Morning'
    elif h >= 10 & h <= 14:
        return 'Peak'
    elif h > 14:
        return 'Afternoon'
    else:
        return 'Other'
```

```python
df
```

```python
df['APPT_DTTM'] = pd.to_datetime(df['APPT_DTTM'])
```

```python
df['TIME_OF_DAY'] = df['APPT_DTTM'].apply(getTimeOfDay)
```
