This notebook creates a CSV that contains only the duplicate rows. This was to help me look at the duplicates and decide how I wanted to proceed with cleaning them.

```python
import pandas as pd
from functions import convertTime
from datetime import datetime, date
```

First a create a dataframe of duplicates through the duplicate subset of the original dataframe.  
Then, I export the duplicate dataframe to a CSV, ready for review.

```python
duplicates = df.duplicated(subset='INPATIENT_DATA_ID_x', keep=False)

df[duplicates].to_csv('duplicateData.csv', index=False)
```
